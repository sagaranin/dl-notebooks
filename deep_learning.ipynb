```python
weights = [0.1, 0.2, 0.3]

def w_sum(a, b):
    assert(len(a) == len(b))
    output = 0
    for i in range(len(a)):
        output += a[i] * b[i]
    return output

def neural_network(inputs, weights):
    prediction = w_sum(inputs, weights)
    return prediction
```

```python
weight = 0.12

lr = 0.01

def neural_network(inputs, weights):
    prediction = inputs * weights
    return prediction

number_of_toes = [8.5]
win_or_lose_binary = [1]
input = number_of_toes[0]
true = win_or_lose_binary[0]

pred = neural_network(input, weight)

error = (pred - true) ** 2
print(error)

p_up = neural_network(input, weight) + lr
p_dn = neural_network(input, weight) - lr

e_up = (p_up - true) ** 2
e_dn = (p_dn - true) ** 2
print(f"e_up {e_up}, e_dn {e_dn}")

```

```python
from math import sqrt

weight = 0.5
input = 0.5
goal_prediction = 0.8

step_amount = 0.2

for i in range(1201):
    prediction = weight * input
    error = (prediction - goal_prediction) ** 2
    print(f"Error: {error}, prediction: {prediction}")
    
    up_prediction = input * (weight + step_amount)
    up_error = error = (up_prediction - goal_prediction) ** 2
    
    down_prediction = input * (weight - step_amount)
    down_error = error = (down_prediction - goal_prediction) ** 2
    
    if (down_error < up_error):
        weight -= step_amount
    
    if (up_error < down_error):
        weight += step_amount
```

```python
weight = 0.5
input = 0.5
goal_pred = 0.8


for i in range(1201):
    pred = weight * input
    error = (pred - goal_pred) ** 2
    direction_and_amount = (pred - goal_pred) * input
    weight -= direction_and_amount
    
    print(f"Error: {error},\tprediction: {pred},\tclear_err {(pred - goal_pred)},\tdaa {direction_and_amount}")
        
```

```python
weight = 1.6
input = 0.5
goal_pred = 0.8

alpha = 1.1

for i in range(1):
    pred = weight * input
    error = (pred - goal_pred) ** 2
    delta = (pred - goal_pred)
    weight_delta = input * delta
    weight -= weight_delta * alpha
    
    print(f"Error: {error},\tprediction: {pred},\tclear_err {(pred - goal_pred)},\tweight_delta {weight_delta}")
```
